## View data

Use the same syntax in the first week to view some of the data

normalize first, and view them

In general, if we download the dataset from Kaggle or some shit website, there's probably some folder called "train", "test", "valid" to make sure your model works fine

````
np.random.seed(2020) #Whenever I run this code, I can get the same result everytime

data = ImageDataBunch.from_folder(path, 
                                  train = ".", #the training set is in the current folder
                                  valid_pct = 0.2, #we set aside 20% of the data as valid data
                                  ds_tfms = get_transforms(), 
                                  size = 224, 
                                  num_workers = 4).normalize(imagenet_stats)

#If things goes wrong, we can get size into 224
````

## Train Model

````
learn = create_cnn(data, models.resnet34, metrics = error_rate)
````

Run 4 epoches and see how it goes

````
learn.fit_one_cycle(4) #4 epoches
````

That is.....awful

Save the first stage model

````
learn.save("stage-1")
````

Unfreeze the model

````
learn.unfreeze()
````

We run the learning rate find

````
learn.lr_find()
````

And plot it

````
learn.recorder.plot()
````

The most important thing in learning rate finder, we are looking for the strongest down for the slope, i.e., the highest slope change.

So in this case, is around 10^-4 to 10^-3

If there's several slopes that looks like the learning rate you are finding, why not try both of them?


````
learn.fit_one_cycle(12, max_lr = slice(1e-4, 1e-3))
````

13% error rate look a little better.

Let us save the model

````
learn.save("stage-2")
````

## Interpretation

````
learn.load("stage-2")
````

````
interp = ClassificationInterpretation.from_learner(learn)
````

Let's see how the classifier going in validation data

````
interp.plot_confusion_matrix()
````



That is around 95% accuracy.

So this model work out not that bad, but not that good either.

And create a ImageDataBunch, but this time we are not going to create from full folder that full of images, this is going to grab one single image at a time.

So we are not gonna passing it any data, we only pass a path. The reason why we do this is that it knows where to load our model from.



So this image data bunch have no image at all, it knows how to transform the new image that we trained with so you can do the inference.

You only have to run this chunk once from startup.

````
classes = data.classes

data2 = ImageDataBunch.single_from_classes(path, classes, ds_tfms = get_transforms(), size = 224).normalize(imagenet_stats)
learn = create_cnn(data2, models.resnet34)
learn.load("stage-2")
````
